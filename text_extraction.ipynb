```python
import pandas as pd

df = pd.read_excel("RAA Data 2020.xlsx")
df
```

```python
combined_medic_notes = " ".join(map(str, df.loc[:, "Medic Notes"]))
combined_narrative = " ".join(map(str, df.loc[:, "Narrative"]))

print(f"Extract from combined medic notes: {combined_medic_notes[0:2000]} + '\n'")
print(f"Extract from combined narrative: {combined_narrative[0:2000]}")
```

```python
from num2words import num2words

def number_to_text_conversion(string):
    string_arr = string.split(" ")
    new_string_arr = []
    for token in string_arr:
        new_token = token
        if token.isnumeric():
            new_token = num2words(token)
        new_string_arr.append(new_token)
    return " ".join(new_string_arr)
```

```python
import re
combined_medic_notes = combined_medic_notes.replace("/", " ")# convert slashes into spaces
combined_medic_notes = re.sub(r'[^\w\s]', ' ', combined_medic_notes) # remove punctuation
combined_medic_notes = number_to_text_conversion(combined_medic_notes)# convert all numbers to text representations (i.e. 69 --> sixty nine)
print(combined_medic_notes[0:1000])
```
